# Data Visualization

### Importing Modules

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Importing our Dataset 

```python
data=pd.read_csv('EGG.csv')
```

```python
data.head()
```

```python
del data['id']
```

```python
data.shape
```

```python
df=data.dropna()
```

### Ploting Hemoglobin for Its overall distribution in the dataset

```python
plt.figure(figsize=(6,6))
plt.hist(df['hemo'])
plt.xlabel('Hemoglobin')
plt.ylabel('Counts')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.hist(df['fasting'])
plt.xlabel('Fasting')
plt.ylabel('Counts')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.hist(df['insulin'])
plt.xlabel('Insulin')
plt.ylabel('Counts')
plt.title('INSULIN DISTRIBUTION')
plt.show()
```

```python
def Contract(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(6,6), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 18)

Contract(df, 'classification')
```

```python
def Contract(df, variable):
    dataframe_pie = df[variable].value_counts()
    ax = dataframe_pie.plot.pie(figsize=(6,6), autopct='%1.2f%%', fontsize = 12)
    ax.set_title(variable + ' \n', fontsize = 10)

Contract(df, 'cpm')
```

```python
plt.figure(figsize=(6,6))
plt.hist(df['postprandial'])
plt.xlabel('Postprandial')
plt.ylabel('Counts')
plt.title('POST PRANDIAL DISTRIBUTION')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.hist(df['final'])
plt.xlabel('final')
plt.ylabel('Counts')
plt.show()
```
